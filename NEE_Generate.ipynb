```python
import pandas as pd
import numpy as np
import os 
import matplotlib.pyplot as plt
from alpha_vantage.timeseries import TimeSeries

%matplotlib inline

from datetime import datetime
import talib

from typing import TypeVar, Callable, Sequence
from functools import reduce
T = TypeVar('T')


from pylab import rcParams

import seaborn as sns
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import cross_val_score, train_test_split, GridSearchCV
from sklearn.feature_selection import RFECV, SelectFromModel, SelectKBest
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from tqdm import tqdm_notebook
from sklearn.preprocessing import StandardScaler
from sklearn import metrics

from keras.optimizers import RMSprop, SGD, Adam
from keras.layers import LSTM, Dense, Dropout
from keras.models import Sequential
%matplotlib inline

ALPHA_VANTAGE_API_KEY = '99RG1ZGZNWB4KLMX'
ts = TimeSeries(key= ALPHA_VANTAGE_API_KEY, output_format='pandas')

import pandas as pd
import pandas_datareader.data as web   
import datetime

start = datetime.datetime(2008,1,2)
end = datetime.datetime(2021,2,24)

#aae = web.DataReader("AAE", "yahoo", start, end)
#aae.to_csv('NEE.csv',index=False,header=True)

NEE = web.DataReader("NEE", "yahoo", start, end)
NEE.to_csv('NEE.csv',index=True,header=True)


NEE['Close'].plot(figsize=(10, 7))
plt.title("Netflix Stock Price", fontsize=17)
plt.ylabel('Price', fontsize=14)
plt.xlabel('Time', fontsize=14)
plt.grid(which="major", color='k', linestyle='-.', linewidth=0.5)
plt.show()
#

####
#
####

QQQ, QQQ_info = ts.get_daily('QQQ', outputsize='full') 
QQQ = QQQ.rename(columns={'1. open' : 'Open', '2. high': 'High', '3. low':'Low', '4. close': 'QQQ_Close', '5. volume': 'Volume' })
QQQ = QQQ.rename_axis(['Date'])
QQQ = QQQ.drop(columns=['Open', 'High', 'Low', 'Volume'])

QQQ = QQQ.sort_index(ascending=True, axis=0)
#slicing the data for 15 years from '2004-01-02' to today
QQQ = QQQ.loc['2008-01-02':]
QQQ

QQQ['QQQ(t-1)'] = QQQ.QQQ_Close.shift(periods=1)
QQQ['QQQ(t-2)'] =  QQQ.QQQ_Close.shift(periods=2)
QQQ['QQQ(t-5)'] =  QQQ.QQQ_Close.shift(periods=5)
QQQ


QQQ['QQQ_MA10'] = QQQ.QQQ_Close.rolling(window=10).mean()
#QQQ['QQQ_MA10_t'] = QQQ.QQQ_ClosePrev1.rolling(window=10).mean()
QQQ['QQQ_MA20'] = QQQ.QQQ_Close.rolling(window=20).mean()
QQQ['QQQ_MA50'] = QQQ.QQQ_Close.rolling(window=50).mean()



NEE['MA_20'] = NEE.Close.rolling(window=20).mean()
NEE['SD20'] = NEE.Close.rolling(window=20).std()
NEE['Upper_Band'] = NEE.Close.rolling(window=20).mean() + (NEE['SD20']*2)
NEE['Lower_Band'] = NEE.Close.rolling(window=20).mean() - (NEE['SD20']*2)
NEE.tail()

NEE[['Close', 'MA_20', 'Upper_Band', 'Lower_Band']].plot(figsize=(12,6))
plt.title('20 Day Bollinger Band for NEE')
plt.ylabel('Price (USD)')
plt.show();

NEE['NEE_Close(t-1)'] = NEE.Close.shift(periods=1)
NEE['NEE_Close(t-2)'] = NEE.Close.shift(periods=2)
NEE['NEE_Close(t-5)'] = NEE.Close.shift(periods=5)
NEE['NEE_Close(t-10)'] = NEE.Close.shift(periods=10)
NEE['NEE_Open(t-1)'] = NEE.Open.shift(periods=1)


NEE['MA5'] = NEE.Close.rolling(window=5).mean()
NEE['MA10'] = NEE.Close.rolling(window=10).mean()
NEE['MA20'] = NEE.Close.rolling(window=20).mean()
NEE['MA50'] = NEE.Close.rolling(window=50).mean()
NEE['MA200'] = NEE.Close.rolling(window=200).mean()

NEE[['Close', 'MA20', 'MA200', 'MA50']].plot()
plt.show()

NEE['EMA_12'] = NEE.Close.ewm(span=12, adjust=False).mean()
NEE['EMA_26'] = NEE.Close.ewm(span=26, adjust=False).mean()
NEE['MACD'] = NEE['EMA_12'] - NEE['EMA_26']

NEE['MACD_EMA'] = NEE.MACD.ewm(span=9, adjust=False).mean()

NEE[['MACD', 'MACD_EMA']].plot()
plt.show()




NEE['EMA10'] = NEE.Close.ewm(span=5, adjust=False).mean().fillna(0)
NEE['EMA20'] = NEE.Close.ewm(span=5, adjust=False).mean().fillna(0)
NEE['EMA50'] = NEE.Close.ewm(span=5, adjust=False).mean().fillna(0)
NEE['EMA100'] = NEE.Close.ewm(span=5, adjust=False).mean().fillna(0)
NEE['EMA200'] = NEE.Close.ewm(span=5, adjust=False).mean().fillna(0)




import talib

NEE['ATR'] = talib.ATR(NEE['High'].values, NEE['Low'].values, NEE['Close'].values, timeperiod=14)
NEE['ADX'] = talib.ADX(NEE['High'], NEE['Low'], NEE['Close'], timeperiod=14)
tp = (NEE['High'] + NEE['Low'] + NEE['Close']) /3
ma = tp/20 
md = (tp-ma)/20
NEE['CCI'] = (tp-ma)/(0.015 * md)

NEE['ROC'] = ((NEE['Close'] - NEE['Close'].shift(10)) / (NEE['Close'].shift(10)))*100

NEE['RSI'] = talib.RSI(NEE.Close.values, timeperiod=14)

NEE['William%R'] = talib.WILLR(NEE.High.values, NEE.Low.values, NEE.Close.values, 14) 

NEE['SO%K'] = ((NEE.Close - NEE.Low.rolling(window=14).min()) / (NEE.High.rolling(window=14).max() - NEE.Low.rolling(window=14).min())) * 100

NEE['per_change'] = NEE.Close.pct_change()
NEE['STD5'] = NEE.per_change.rolling(window=5).std()


SnP = web.DataReader("^GSPC", "yahoo", start, end)
SnP.to_csv('INX.csv',index=True,header=True)


#SnP, SnP_info = ts.get_daily('INX', outputsize='full') 
SnP = SnP.rename(columns={'1. open' : 'Open', '2. high': 'High', '3. low':'Low', 'Close': 'SnP_Close', '5. volume': 'Volume' })
SnP = SnP.rename_axis(['Date'])
SnP = SnP.drop(columns=['Open', 'High', 'Low', 'Volume'])

#sorting index
SnP = SnP.sort_index(ascending=True, axis=0)
#slicing the data for 15 years from '2004-01-02' to today
SnP = SnP.loc['2008-01-02':]
SnP

SnP['SnP(t-1))'] = SnP.SnP_Close.shift(periods=1)
SnP['SnP(t-5)'] =  SnP.SnP_Close.shift(periods=5)



NEE = NEE.merge(QQQ, left_index=True, right_index=True)
NEE = NEE.merge(SnP, left_index=True, right_index=True)
NEE = NEE.drop(columns=['MA_20', 'per_change', 'EMA_12', 'EMA_26'])
NEE['ForceIndex1'] = NEE.Close.diff(1) * NEE.Volume
NEE['ForceIndex20'] = NEE.Close.diff(20) * NEE.Volume
NEE['NEE_Close(t+1)'] = NEE.Close.shift(-1)

NEE = NEE.dropna()
NEE['Date_Col'] = NEE.index


from datetime import datetime

def extract_date_features(date_val):
    
    Day = date_val.day
    DayofWeek = date_val.dayofweek
    Dayofyear = date_val.dayofyear
    Week = date_val.week
    Is_month_end = date_val.is_month_end.real
    Is_month_start = date_val.is_month_start.real
    Is_quarter_end = date_val.is_quarter_end.real
    Is_quarter_start = date_val.is_quarter_start.real
    Is_year_end = date_val.is_year_end.real
    Is_year_start = date_val.is_year_start.real
    Is_leap_year = date_val.is_leap_year.real
    Year = date_val.year
    Month = date_val.month
    
    return Day, DayofWeek, Dayofyear, Week, Is_month_end, Is_month_start, Is_quarter_end, Is_quarter_start, Is_year_end, Is_year_start, Is_leap_year, Year, Month




funct = lambda x: pd.Series(extract_date_features(x))
NEE[['Day', 'DayofWeek', 'DayofYear', 'Week', 'Is_month_end', 'Is_month_start',
      'Is_quarter_end', 'Is_quarter_start', 'Is_year_end', 'Is_year_start', 'Is_leap_year', 'Year', 'Month']] = NEE.Date_Col.apply(funct)



DJIA = web.DataReader("^DJI", "yahoo", start, end)
DJIA.to_csv('DJIA.csv',index=True,header=True)

#DJIA, DJIA_info = ts.get_daily('^DJI', outputsize='full') 
DJIA = DJIA.rename(columns={'Close': 'DJIA_Close' })
DJIA = DJIA.rename_axis(['Date'])
DJIA = DJIA.drop(columns=['Open', 'High', 'Low', 'Volume','Adj Close'])

#sorting index
DJIA = DJIA.sort_index(ascending=True, axis=0)
        #slicing the data for 15 years from '2004-01-02' to today

DJIA['DJIA(t-1))'] = DJIA.DJIA_Close.shift(periods=1)
DJIA['DJIA(t-5)'] =  DJIA.DJIA_Close.shift(periods=5)

NEE = NEE.merge(DJIA, left_index=True, right_index=True)

NEE.to_csv('NEE.csv')
```
